```python
import cv2
from pytesseract import pytesseract
from pytesseract import Output
from google.colab.patches import cv2_imshow
```

```python
#Put image location Here
IMAGE_PATH = '/content/drive/MyDrive/OCR /shakespeare.JPG'
```

```python
#Check Which Words Are recognised

img = cv2.imread(IMAGE_PATH)
height, width, c = img.shape

image_data = pytesseract.image_to_data(img, output_type=Output.DICT)

for i, word in enumerate(image_data['text']):
  if word != "":
    x,y,w,h = image_data['left'][i],image_data['top'][i],image_data['width'][i],image_data['height'][i]
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),3)
    cv2.putText(img, word, (x,y-16), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,255),1)


cv2_imshow(img)
```

```python
#Printing text
img = cv2.imread(IMAGE_PATH)
words_in_image= pytesseract.image_to_string(img)

print(words_in_image)
```
