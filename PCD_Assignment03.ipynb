```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#image 
img = cv2.imread('/content/drive/My Drive/BNW.jpg')

#kernel
kernel = np.ones((5,5),np.uint8)


#morphologies
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

print("OG image: ")
cv2_imshow(img)
print("\n")

print("Erosion image: ")
cv2_imshow(erosion)
print("\n")

print("Dilation image: ")
cv2_imshow(dilation)
print("\n")

print("Opening image: ")
cv2_imshow(opening)
print("\n")

print("Closing image: ")
cv2_imshow(closing)
```
